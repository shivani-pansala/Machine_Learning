```python
import pandas as pd
from sklearn import linear_model,model_selection
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error

#load data
data=pd.read_csv("housing_data.csv")
# print(len(x))
# print(len(y))

#split the dataset
x=np.array(data.age).reshape(-1,1)
y=np.array(data.price).reshape(-1,1)

#load the model
linear_regression_model=linear_model.LinearRegression()

xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x,y,test_size=0.2)

#train the model
linear_regression_model.fit(xtrain,ytrain)

ytest_result=linear_regression_model.predict(xtest)
#print(xtest)
#print(ytest)

plt.scatter(xtest,ytest,color="red")
plt.plot(xtest,ytest_result,color="blue")
plt.show()

print("mean absolute error:" ,mean_absolute_error(ytest,ytest_result))
print("mean square error:" ,mean_squared_error(ytest,ytest_result))
print("mean absolute percentage error:" ,mean_absolute_percentage_error(ytest,ytest_result))
```
