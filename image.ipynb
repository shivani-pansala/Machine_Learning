```python
#image read
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread("girl.jpg")
newmg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(newmg),plt.axis("off")
```

```python
r1=np.array([[-1,-1,-1],[-1,1,-1],[-1,-1,-1]])
filters=cv2.filter2D(img,-3,r1)
plt.imshow(filters),plt.axis("off")
```

```python
newimg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
BLUR=cv2.blur(newimg,(10,10))
plt.subplot(1,2,1),plt.imshow(newimg),plt.axis("off"),plt.title("original image")
plt.subplot(1,2,2),plt.imshow(BLUR),plt.axis("off"),plt.title("Blur image")
```

```python
#gaussian and median blur

import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("girl.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
blr = cv2.blur(img,(190,19))
gaussian = cv2.GaussianBlur(img,(59,59),0)
medians = cv2.medianBlur(img,19)

plt.subplot(1,4,1),plt.imshow(blr),plt.axis("off"),plt.title("Blur ")
plt.subplot(1,4,2),plt.imshow(gaussian),plt.axis("off"),plt.title("Gaussian Blur ")
plt.subplot(1,4,3),plt.imshow(medians),plt.axis("off"),plt.title("Median Blur ")
plt.subplot(1,4,4),plt.imshow(img),plt.axis("off"),plt.title(" Original ")
```

# sharpening


```python
img = cv2.imread("girl.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
k1 = np.array([[-1,-1,-1],
               [-1,9,-1],
               [-1,-1,-1]])
k2=np.array([[1,1,1],
               [1,-7,1],
               [1,1,1]])
sharp=cv2.filter2D(img,-1,k1)
sharpblur = cv2.filter2D(img,-1,k2)

plt.subplot(1,2,1),plt.imshow(sharp),plt.axis("off"),plt.title("sharpening image")
plt.subplot(1,2,2),plt.imshow(sharpblur),plt.axis("off"),plt.title("sharpening blur image")
```

```python
k1 = np.array([[0,-1,-1],
               [1,0,-1],
               [1,1,0]])
k2 = np.array([[-1,-1,0],
               [-1,0,1],
               [0,1,1]])
k3 = np.array([[0,0,-1],
               [0,0,0],
               [-1,0,0]])
k4 = np.array([[1,0,0],
               [0,0,0],
               [0,0,1]])

emboss1 = cv2.filter2D(img,-1,k1)+128
emboss2 = cv2.filter2D(img,-1,k2)+128
emboss3 = cv2.filter2D(img,-1,k3)+128
emboss4 = cv2.filter2D(img,-1,k4)+128


plt.subplot(1,4,1),plt.imshow(emboss1),plt.axis("off"),plt.title("embossing")
plt.subplot(1,4,2),plt.imshow(emboss2),plt.axis("off"),plt.title("embossing1")
plt.subplot(1,4,3),plt.imshow(emboss3),plt.axis("off"),plt.title("embossing2")
plt.subplot(1,4,4),plt.imshow(emboss4),plt.axis("off"),plt.title("embossing3")
```

# sobel

```python
img = cv2.imread("girl.jpg")
new = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#horizontal
sobelsX = cv2.Sobel(img,cv2.CV_64F,1,0,5)
#vertical
sobelsY = cv2.Sobel(img,cv2.CV_64F,0,1,5)
plt.subplot(1,3,1),plt.imshow(new),plt.axis("off"),plt.title("original")
plt.subplot(1,3,2),plt.imshow(sobelsX),plt.axis("off"),plt.title("horizontal")
plt.subplot(1,3,3),plt.imshow(sobelsY),plt.axis("off"),plt.title("vertical")


```

# LAPLACIAN 

```python
img = cv2.imread("girl.jpg")
newimg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
lap =cv2.Laplacian(newimg,cv2.CV_64F)
plt.imshow(lap,cmap="gray"),plt.axis("off")
```

# Canny 

```python
cany=cv2.Canny(newimg,50,150)
plt.subplot(1,2,1),plt.imshow(newimg),plt.axis("off"),plt.title("original")
plt.subplot(1,2,2),plt.imshow(cany),plt.axis("off"),plt.title("canny")
```

# Erosion & Dilation

```python
k1=np.ones((5,5))
eros=cv2.erode(newimg,k1,iterations=1)
dial=cv2.dilate(newimg,k1,iterations=5)

plt.subplot(1,3,1),plt.imshow(newimg),plt.axis("off"),plt.title("original")
plt.subplot(1,3,2),plt.imshow(eros),plt.axis("off"),plt.title("Eroded")
plt.subplot(1,3,3),plt.imshow(dial),plt.axis("off"),plt.title("Dialation")
```

# color contrast

```python
yuv=cv2.cvtColor(newimg,cv2.COLOR_BGR2YUV)
yuv[:,:,0]=cv2.equalizeHist(yuv[:,:,0])
finalimg=cv2.cvtColor(yuv,cv2.COLOR_YUV2BGR)
plt.subplot(1,2,1),plt.imshow(newimg),plt.axis("off"),plt.title("original")
plt.subplot(1,2,2),plt.imshow(finalimg),plt.axis("off"),plt.title("color contrast")
```

# horizontal and vertical blur

```python
k2=np.zeros((50,50))
k2[int(50/2),:]=np.ones(50)
k2=k2/50

k3=np.zeros((50,50))
k3[:,int(50/2)]=np.ones(50)
k3=k3/50

horblr=cv2.filter2D(newimg,-1,k2)
verblr=cv2.filter2D(newimg,-1,k3)

plt.subplot(1,3,1),plt.imshow(newimg),plt.axis("off"),plt.title("original")
plt.subplot(1,3,2),plt.imshow(horblr),plt.axis("off"),plt.title("horizontal")
plt.subplot(1,3,3),plt.imshow(verblr),plt.axis("off"),plt.title("vertical")
```
